# This is the basic idea behind the final Capstone project:

 The main idea is to help people who want to open a new restaurant in Canda specially Tornto to choose the best place for their project. 
By exploring the recent positions of the current restaurants and leverage the position of the restaurants in Tornto and their distributions in the city, we can find the best place.

# Data Description 

I will depend on the data about the recent Locations of Tornto restaurant, the degree of quality, customer satisfaction and also the locations of the least restaurants and high population rates in the city. By using the Foursquare database and rating system, we can reach our destination.
